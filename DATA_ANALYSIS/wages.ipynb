```python
df.columns = [
    'Governorates', 'Total_Avg_Work_Hours', 'Total_Avg_Salary', 'Total_Count',
    'Female_Avg_Work_Hours', 'Female_Avg_Salary', 'Female_Count',
    'Male_Avg_Work_Hours', 'Male_Avg_Salary', 'Male_Count'
]

# Convert relevant columns to numeric types
df['Total_Avg_Work_Hours'] = pd.to_numeric(df['Total_Avg_Work_Hours'], errors='coerce')
df['Total_Avg_Salary'] = pd.to_numeric(df['Total_Avg_Salary'], errors='coerce')
df['Total_Count'] = pd.to_numeric(df['Total_Count'], errors='coerce')
df['Female_Avg_Work_Hours'] = pd.to_numeric(df['Female_Avg_Work_Hours'], errors='coerce')
df['Female_Avg_Salary'] = pd.to_numeric(df['Female_Avg_Salary'], errors='coerce')
df['Female_Count'] = pd.to_numeric(df['Female_Count'], errors='coerce')
df['Male_Avg_Work_Hours'] = pd.to_numeric(df['Male_Avg_Work_Hours'], errors='coerce')
df['Male_Avg_Salary'] = pd.to_numeric(df['Male_Avg_Salary'], errors='coerce')
df['Male_Count'] = pd.to_numeric(df['Male_Count'], errors='coerce')
```

```python
avg_salaries_by_governorate = df.groupby('Governorates')[['Female_Avg_Salary', 'Male_Avg_Salary']].mean()
print(avg_salaries_by_governorate)
```

```python
avg_female_work_hours = df.groupby('Governorates')['Female_Avg_Work_Hours'].mean()
avg_male_work_hours = df.groupby('Governorates')['Male_Avg_Work_Hours'].mean()
```

```python
avg_female_work_hours = df.groupby('Governorates')['Female_Avg_Work_Hours'].mean()
avg_male_work_hours = df.groupby('Governorates')['Male_Avg_Work_Hours'].mean()

# Combine the results into a single DataFrame
avg_work_hours_by_governorate = pd.DataFrame({
    'Female_Avg_Work_Hours': avg_female_work_hours,
    'Male_Avg_Work_Hours': avg_male_work_hours
})
```

```python
print(avg_work_hours_by_governorate)
```

```python
avg_total_work_hours = df.groupby('Governorates')['Total_Avg_Work_Hours'].mean()
avg_total_salaries = df.groupby('Governorates')['Total_Avg_Salary'].mean()
```

```python
avg_total_by_governorate = pd.DataFrame({
    'Total_Avg_Work_Hours': avg_total_work_hours,
    'Total_Avg_Salary': avg_total_salaries
})
```

```python
print("Average Total Work Hours and Salaries by Governorate:")
print(avg_total_by_governorate)
```

```python
# sort
df = df.dropna()
df = df.sort_values(by="Total_Avg_Work_Hours",ascending=False)
```

```python
# Create a line chart for average work hours by gender in each governorate
plt.figure(figsize=(12, 6))

# Plot the line for Female Average Work Hours
plt.plot(df['Governorates'], df['Female_Avg_Work_Hours'], marker='o', color='skyblue', label='Female Average Work Hours')

# Plot the line for Male Average Work Hours
plt.plot(df['Governorates'], df['Male_Avg_Work_Hours'], marker='o', color='salmon', label='Male Average Work Hours')

plt.xticks(rotation=90)
plt.xlabel('Governorates')
plt.ylabel('Average Work Hours')
plt.title('Average Work Hours by Gender in Each Governorate')
plt.legend()
plt.grid()

plt.show()
# Create visualizations for average salaries by gender in each governorate
```

```python
df = df.sort_values(by="Total_Avg_Salary",ascending = False)
plt.figure(figsize=(12, 6))

# Plot the line for Female Average Salary
plt.plot(df['Governorates'], df['Female_Avg_Salary'], marker='o', color='skyblue', label='Female Average Salary')

# Plot the line for Male Average Salary
plt.plot(df['Governorates'], df['Male_Avg_Salary'], marker='o', color='salmon', label='Male Average Salary')

plt.xticks(rotation=90)
plt.xlabel('Governorates')
plt.ylabel('Average Salary')
plt.title('Average Salary by Gender in Each Governorate')
plt.legend()
plt.grid()

plt.show()
```
