```python
import pandas as pd
```

```python
public_teachers_2022 = pd.read_excel(r"D:\Data Analysis\Uncleaned Data\Public_edu_22.xlsx")
public_teachers_2022.head(50)
```

```python
public_teachers_2022 = public_teachers_2022.drop(labels="Unnamed: 0",axis=1)
```

```python
public_teachers_2022 = public_teachers_2022.drop(public_teachers_2022.index[:1])
```

```python
public_teachers_2022 = public_teachers_2022[public_teachers_2022.columns[::-1]]
```

```python
labels = ['Governorate', 'male_teachers_number', 'female_teachers_number', 'teachers_numbers','male_students_number', 'female_students_number', 'students_numbers']
public_teachers_2022.columns = labels
```

```python
public_teachers_2022.head(50)
```

```python
public_teachers_2022.isnull().sum()
```

```python
public_teachers_2022.info()
```

```python
numerical_varialbes = ['male_teachers_number', 'female_teachers_number', 'teachers_numbers','male_students_number', 'female_students_number', 'students_numbers']
for column in numerical_varialbes:
    public_teachers_2022[column] = public_teachers_2022[column].astype(str).str.replace(' ', '').str.replace(',', '')
    public_teachers_2022[column] = public_teachers_2022[column].fillna(public_teachers_2022[column].mean()).astype(int)
```

```python
public_teachers_2022.info()
```

```python
public_teachers_2022.drop(['teachers_numbers', 'students_numbers'], axis=1, inplace=True)
public_teachers_2022['teachers_numbers'] = public_teachers_2022['male_teachers_number'] + public_teachers_2022['female_teachers_number']
public_teachers_2022['students_numbers'] = public_teachers_2022['male_students_number'] + public_teachers_2022['female_students_number']
```
