```python
new_df = pd.DataFrame(columns=['Governorate', 'Place_of_residence', 'Gender', 'Total', 'High_Education', 'Medium_Education', 'Technical_Education', 'Secondary_Education', 'Below_Secondary_Education', 'Illiterate', 'Semiliterate'])

# Find all the occurrences of 'Total' in the 'Place_of_residence' column
total_rows = df[df['Place_of_residence'] == 'Total']

# Iterate over the unique governorates
for i, governorate in enumerate(df['Governorate'].unique()):
    # Filter the rows for the current governorate
    governorate_rows = df[df['Governorate'] == governorate]
    #print(governorate_rows)

    # Get the corresponding 'Total' row for the current governorate
    if len(total_rows) > i :
        total_row_index = total_rows.index[i]

        # Extract the place of residence, gender, total value, and education-related values
        place_of_residence = df.loc[total_row_index, 'Place_of_residence']
        gender = df.loc[total_row_index + 1, 'Gender']
        total_value = df.loc[total_row_index + 1, 'Total']
        high_education = df.loc[total_row_index + 1, 'High_Education']
        medium_education = df.loc[total_row_index + 1, 'Medium_Education']
        technical_education = df.loc[total_row_index + 1, 'Technical_Education']
        secondary_education = df.loc[total_row_index + 1, 'Secondary_Education']
        below_secondary_education = df.loc[total_row_index + 1, 'Below_Secondary_Education']
        illiterate = df.loc[total_row_index + 1, 'Illiterate']
        semiliterate = df.loc[total_row_index + 1, 'Semiliterate']

        # Create a new row dictionary
        new_row = {
            'Governorate': governorate,
            'Place_of_residence': place_of_residence,
            'Gender': gender,
            'Total': total_value,
            'High_Education': high_education,
            'Medium_Education': medium_education,
            'Technical_Education': technical_education,
            'Secondary_Education': secondary_education,
            'Below_Secondary_Education': below_secondary_education,
            'Illiterate': illiterate,
            'Semiliterate': semiliterate
        }

        # Append the new row to the new DataFrame
        new_df = pd.concat([new_df, pd.DataFrame([new_row])], ignore_index=True)
# Print the new DataFrame
new_df.to_excel("last.xlsx", index=False)

print(new_df)
```
