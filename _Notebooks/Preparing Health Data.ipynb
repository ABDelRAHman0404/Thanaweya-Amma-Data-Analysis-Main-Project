```python
hospitals = pd.read_csv(r"D:\Data Analysis\Uncleaned Data\hospitalsbygovernrate.csv")
hospitals.head()
```

```python
hospitals.isnull().sum()
```

```python
hospitals = hospitals.drop(labels="Total",axis=1)
```

```python
hospitals.columns = ['Governorate', 'Population', 'Hospitals_numbers', 'Beds_numbers', 'Intensive_Care_Beds_numbers', 'Incubators_numbers', 'Dialysis_Devices_numbers', 'Respirators_numbers']
```

```python
hospitals['Respirators_numbers'] = hospitals['Respirators_numbers'].fillna(hospitals['Respirators_numbers'].mean())
hospitals['Incubators_numbers'] = hospitals['Incubators_numbers'].fillna(hospitals['Incubators_numbers'].mean())
hospitals['Dialysis_Devices_numbers'] = hospitals['Dialysis_Devices_numbers'].fillna(hospitals['Dialysis_Devices_numbers'].mean())
```

```python
hospitals['Population'] = hospitals['Population'].str.replace(',', '').astype(int)
hospitals['Hospitals_numbers'] = hospitals['Hospitals_numbers'].astype(int)
hospitals['Beds_numbers'] = hospitals['Beds_numbers'].astype(int)
hospitals['Intensive_Care_Beds_numbers'] = hospitals['Intensive_Care_Beds_numbers'].astype(int)
hospitals['Incubators_numbers'] = hospitals['Incubators_numbers'].astype(int)
hospitals['Dialysis_Devices_numbers'] = hospitals['Dialysis_Devices_numbers'].astype(int)
hospitals['Respirators_numbers'] = hospitals['Respirators_numbers'].astype(int)
```

```python
visitors['Visitors_numbers'] = visitors['Visitors_numbers'].astype(int)
```

```python
hospitals.head()
```

```python
hospitals = hospitals.sort_values(by = ['Governorate'])
visitors = visitors.sort_values(by = ['Governorate'])
```

```python
health = pd.concat([hospitals, visitors[["Visitors_numbers"]]], axis=1)
```

```python
health.info()
```

```python
health.shape
```

```python
health = health.drop(health.index[-1])
```

```python
health['Hospital-percentage'] = (health['Hospitals_numbers'] / health['Population']) * 100
health['Beds_percentage'] = (health['Beds_numbers'] / health['Population']) * 100
health['Intensive_care_beds_percentage'] = (health['Intensive_Care_Beds_numbers'] / health['Population']) * 100
health['Incubators_percentage'] = (health['Incubators_numbers'] / health['Population']) * 100
health['Dialysis_devices_percentage'] = (health['Dialysis_Devices_numbers'] / health['Population']) * 100
health['Respirators_percentage'] = (health['Respirators_numbers'] / health['Population']) * 100
health['Visitors_numbers_percentage'] = (health['Visitors_numbers'] / health['Population']) * 100
```

```python
health.to_csv("D:\Data Analysis\Cleaned Data\health.csv",index = False)
```
